univariate analysis

```python
sns.histplot(laptop['Price'],bins=20,kde= True)
```

```python
sns.boxplot(laptop['Price'])
```

Observation:
The price distribution is positively skewed with most laptops clustered in the lower and mid-price segments. A small number of premium laptops create a long right tail.

ML Insight:
Since the target variable is skewed and contains high-price outliers, logarithmic transformation of the Price column may improve regression model performance.


```python
numerical_cols = [
    'Inches','TouchScreen','Ips','X_res','Y_res',
    'ppi','Dedicated_Gpu','Ram_GB','Weight_kg',
    'SSD','HHD','Total_Storage_GB','Price'
]

categorical_cols = [
    'Company','TypeName','OpSys',
    'Storage_Type','Storage_Category'
]

text_cols = [
    'ScreenResolution','Cpu','Gpu'
]
```

```python
laptop['TouchScreen'].value_counts()
```

```python
laptop['Ram_GB'].value_counts().sort_values()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,3))
    sns.histplot(laptop[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,2))
    sns.boxplot(x=laptop[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

# bivariate analysis

PRICE V/S COMPANY

```python
sns.barplot(
    x='Price',
    y='Company',
    data=laptop,
    order=laptop.groupby('Company')['Price']
                .mean()
                .sort_values()
                .index
)
```

```python
order = laptop.groupby('Company')['Price'].median().sort_values().index

sns.boxplot(
    x='Price',
    y='Company',
    data=laptop,
    order=order
)
```

```python
sns.barplot(y=laptop['Cpu_brand'],x=laptop['Price'],order=laptop.groupby('Cpu_brand')['Price'].mean().sort_values().index,palette='rainbow')
```

```python
sns.barplot(y=laptop['Cpu_brand'],x=laptop['Price'],order=laptop.groupby('Cpu_brand')['Price'].mean().sort_values().index,palette='rainbow',hue=laptop['OpSys'])
```
